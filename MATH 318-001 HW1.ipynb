# Out-of-State Tuition Box Plots

e) Use the `boxplot()` method of college to produce side-by-side boxplots of `Outstate` versus `Private`.

```python
# Create a fig container and an ax plot
fig, ax = plt.subplots(figsize=(10,10))

# Instead of making a new plot, place boxplot inside the ax plot
college.boxplot(column='Outstate', by='Private', ax=ax)

# Rename title and y-axis label for clarity
ax.set_title('Out-of-state Tuition by Private/Public Indicator')
ax.set_ylabel('Out-of-state Tuition ($)')

# Remove the auto-generated pandas title
plt.suptitle("")

# Display the plot
plt.show()
```

(f) Create a new qualitative variable, called Elite, by binning the `Top10perc` variable into two groups based on whether or not the proportion of students coming from the top 10% of their high school classes exceeds 50%.

```python
# Create 'Elite' variable by binning Top10perc
# Since Top10perc is the percentage of students, divide by 100 for the number of students
college['Elite'] = pd.cut(college['Top10perc']/100, [0, 0.5, 1], labels=['No', 'Yes'])
```

Use the `value_counts()` method of `college['Elite']` to see how many elite universities there are.

```python
# Check counts
college['Elite'].value_counts()
```

Finally, use the `boxplot()` method again to produce side-by-side boxplots of `Outstate` versus `Elite`.

```python
# Create a fig container and an ax plot
fig, ax = plt.subplots(figsize=(10,10))

# Instead of making a new plot, place boxplot inside the ax plot
college.boxplot(column='Outstate', by='Elite', ax=ax)

# Rename title and y-axis label for clarity
ax.set_title('Out-of-state Tuition by Elite/ Non-Elite Status')
ax.set_ylabel('Out-of-state Tuition ($)')

# Remove the auto-generated pandas title
plt.suptitle("")

# Display the plot
plt.show()
```

# Histograms with Qualitative Variables

(g) Use the `plot.hist()` method of `college` to produce some histograms with differing numbers of bins for a few of the quantitative variables. The command `plt.subplots(2, 2)` may be useful: it will divide the plot window into four regions so that four plots can be made simultaneously.
```python
# Create a 2x2 grid
fig, ax = plt.subplots(2, 2, figsize=(10, 10))

# Plot histograms for quantitative variables with different bins on specific axes
college.plot.hist(column='Grad.Rate', ax=ax[0,0], bins=5)
college.plot.hist(column='Outstate', ax=ax[0,1], bins=10)
college.plot.hist(column='S.F.Ratio', ax=ax[1,0], bins=15)
college.plot.hist(column='Top10perc', ax=ax[1,1], bins=20)

# Clean up annotations
ax[0,0].set_title('Grad.Rate (5 Bins)')
ax[0,0].set_xlabel('Graduation Rate (%)')
ax[0,1].set_title('Outstate (10 Bins)')
ax[0,1].set_xlabel('Out-of-state Tuition ($)')
ax[1,0].set_title('S.F.Ratio (15 Bins)')
ax[1,0].set_xlabel('Student/faculty Ratio')
ax[1,1].set_title('Top10perc (20 Bins)')
ax[1,1].set_xlabel('New Students in Top 10% of High School Class (%)')
```

```python
# Does S.F.Ratio affect Graduation Rates?
# Create simple scatter plot
plt.scatter(college['S.F.Ratio'], college['Grad.Rate'])
plt.xlabel('Student/faculty Ratio')
plt.ylabel('Graduation Rate')
plt.title('Student.faculty Ratio vs. Graduation Rate')
plt.show()

# Compute the correlation coefficient
correlation = college['S.F.Ratio'].corr(college['Grad.Rate'])
print(f"Correlation between S.F. Ratio and Grad.Rate: {correlation:.2f}")
```
