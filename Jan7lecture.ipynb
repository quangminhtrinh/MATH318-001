# Intro to Plotting
This is handled by `matplotlib` pacakge.

```python
from matplotlib import pyplot as plt
```

## Line Plots

```python
x = np.linspace(0, 10, 100) # 100 points between 0 and 10, inclusive
y = np.sin(x) # sine of each point in x
```

```python
plt.plot(x, y)
plt.show()
```

```python
plt.plot(x, y)
plt.title('Sine Function')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

We can use LaTeX here:

```python
plt.plot(x, y)
plt.title('Sine Function')
plt.xlabel(r'$x$') # raw string to allow for LaTeX formatting
plt.ylabel(r'$y$')
plt.show()
```

```python
plt.plot(x, np.sin(x), label=r'$\sin(x)$')
plt.plot(x, np.cos(x), label=r'$\cos(x)$')
plt.title('Sine and Cosine Functions')
plt.xlabel(r'$x$')
plt.ylabel(r'$y$')
plt.legend()
plt.show()
```

```python
plt.plot(x, np.sin(x), label=r'$\sin(x)$')
plt.plot(x, np.cos(x), label=r'$\cos(x)$')
plt.title('Sine and Cosine Functions')
plt.xlabel(r'$x$')
plt.ylabel(r'$y$')
plt.legend()
plt.savefig('sine_cosine_plot.png') # save the figure to a file
```
